```python
import pandas as pd 

student=pd.read_csv("student_data.csv")
(student)
```

```python
# Print first 7 records
(student.head(7))
```

```python
# Print the column names
(student.columns)
```

```python
# Print all the records in alphabetical order
(student.sort_values(by=['name']))
```

```python
# Print all the records in a way that the students having the highest marks are displayed first
(student.sort_values(by=['Marks1', 'Marks2', 'Marks3', 'Marks4'], ascending=False))
```

```python
# Add a 'percentage' column
student['percentage'] = (student['Marks1'] + student['Marks2'] + student['Marks3'] + student['Marks4']) / 4
```

```python
print(student.columns)
```

```python
# Print the details of students who live in Rajkot
(student[student['city'] == 'Rajkot'])
```

```python
# Print the students who achieve marks < 35
(student[student['Marks1'] < 35])
(student[student['Marks2'] < 35])
(student[student['Marks3'] < 35])
(student[student['Marks4'] < 35])
```

```python
# Print the students whose name starts with 'M'
(student[student['name'].str.startswith('M')])
```

```python
# Find the student having the maximum percentage
(student[student['percentage'] == student['percentage'].max()])
```

```python
# Find the student having the minimum percentage
(student[student['percentage'] == student['percentage'].min()])
```

```python
# Print the students who study in BCA and MCA
student[student['Stream'].isin(['BCA', 'MCA'])]
```
