```python
import cv2
import numpy as np

# Function to blur an image
def blur_image(image):
    blurred = cv2.GaussianBlur(image, (15, 15), 0)
    return blurred

# Function to sharpen an image
def sharpen_image(image):
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    sharpened = cv2.filter2D(image, -1, kernel)
    return sharpened

# Function to emboss an image
def emboss_image(image):
    kernel = np.array([[0,-1,-1], [1,0,-1], [1,1,0]])
    embossed = cv2.filter2D(image, -1, kernel)
    return embossed

# Function to detect edges in an image
def edge_detect(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    return edges

# Function to display menu options
def display_menu():
    print("\nMENU")
    print("1. Blur image")
    print("2. Sharpen image")
    print("3. Emboss image")
    print("4. Detect edges")
    print("5. Exit")
    choice = input("Enter your choice (1/2/3/4/5): ")
    return choice

# Main function to run the program
def main():
    image_path = 'img5.jpg'  # Replace with your image path
    image = cv2.imread(image_path)

    while True:
        choice = display_menu()

        if choice == '1':
            print("\nApplying blur to the image...")
            processed_image = blur_image(image)
            cv2.imshow("Blurred Image", processed_image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

        elif choice == '2':
            print("\nApplying sharpening to the image...")
            processed_image = sharpen_image(image)
            cv2.imshow("Sharpened Image", processed_image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

        elif choice == '3':
            print("\nApplying embossing to the image...")
            processed_image = emboss_image(image)
            cv2.imshow("Embossed Image", processed_image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

        elif choice == '4':
            print("\nDetecting edges in the image...")
            processed_image = edge_detect(image)
            cv2.imshow("Edge Detection", processed_image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

        elif choice == '5':
            print("Exiting program...")
            break

        else:
            print("Invalid choice. Please enter a valid option (1/2/3/4/5).")

if __name__ == "__main__":
    main()
```
